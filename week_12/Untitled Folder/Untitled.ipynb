Q2: join card and customer

```python
card_base.merge(customer_base, how='left', on='Cust_ID')
```

Q3: join fraud and transactions - full

```python
fraud_base.merge(transaction_base, how='outer', on='Transaction_ID')
```

Q4: Join the data containing card details with transaction details so rows from the first which have matching details in the second are returned

```python
transactions_no_dupes = (
    transaction_base.drop_duplicates('Credit_Card_ID', keep='first')
)

card_base.merge(transactions_no_dupes, how='inner', left_on='Card_Number', right_on="Credit_Card_ID")
```

Q6

```python
hat_observations['day'] = hat_observations['day'].astype(str)
hat_observations['month'] = hat_observations['month'].astype(str)
hat_observations['year'] = hat_observations['year'].astype(str)

hat_observations['date'] = hat_observations[['day', 'month', 'year']].agg('-'.join, axis=1)
```

```python
hat_observations
```

```python
counts = hat_observations.groupby('date').size()
counts
```

```python
(
    hat_observations
    .loc[(hat_observations['hat_type'] == 'beret')]
    .drop(columns=['day', 'month', 'year'])
    .groupby('date', as_index=False)
    .agg(observations = ('observation_count', 'sum'))
    .sort_values('observations')
    .tail(1)
)
```
